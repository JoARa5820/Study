# 주성분분석(PCA : Principal Component Analysis)

------

```python
# 다음 그림은 두개의 특성을 가지고 있는 데이터 샘플의 분포에서 두개의 주성분을 나타냄
from IPython.display import Image
Image("PCA_capture.PNG")
```

```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt
```

```python
data = datasets.load_boston()
```

```python
x, y = data['data'], data['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state = 1)
```

```python
scaler = StandardScaler()
x_train_std = scaler.fit_transform(x_train)  # train에는 fit & transform 적용
x_test_std = scaler.transform(x_test)        # test에는 transform만 적용
```

```python
pca = PCA(n_components = 10)  # 추출하고자 하는 주성분의 개수 / 10개의 주성분만 추출
x_train_pca = pca.fit_transform(x_train_std)  # train에는 fit & transform 적용
x_test_pca = pca.transform(x_test_std)        # test에는 transform만 적용
print('원본 입력데이터의 특성 개수 : ',len(x_train[0]))
print('주성분분석이 적용된 입력데이터의 특성 개수 : ',len(x_train_pca[0]))
```

```python
model = KNeighborsRegressor(n_neighbors = 10)
```

```python
model.fit(x_train_pca, y_train)
```

```python
y_predicted = model.predict(x_test_pca)
```

```python
train_score = model.score(x_train_pca, y_train)
test_score = model.score(x_test_pca, y_test)
```

```python
print('학습데이터의 R^2 : {score:.3f}'.format(score = train_score))
```

```python
print('  학습 데이터의 R^2 : ',round(train_score,3))
print('테스트 데이터의 R^2 : ',round(test_score,3))
```

```python
# 실제값과 예측값 사이의 관계를 꺾은 선 그래프로 그림
number_of_sample = 40
plt.plot(range(number_of_sample), y_test[:number_of_sample], label = 'real target')
plt.plot(range(number_of_sample), y_predicted[:number_of_sample], label = 'predicted target')
plt.ylabel('price')
plt.legend()
plt.show()
```

```python
# 판다스의 scatter_matrix 함수를 사용해서 데이터 전체를 한 번에 시각화
import pandas as pd
from pandas.plotting import scatter_matrix

# loading dataset
iris = datasets.load_iris()

# creating data frame for pandas
dataframe = pd.DataFrame(iris['data'], columns=iris['feature_names'])
scatter_matrix(dataframe, c=iris['target'],marker='o', s=10,alpha=.8, figsize=(12,8))
plt.show()
```

### Featrue & Target 확인하기

```python
X = iris.data
y = iris.target
feature_names = iris.feature_names
```

```python
feature_names
```

```python
# 종속변수가 골고루 있는지 확인
df_X = pd.DataFrame(X)
df_Y = pd.DataFrame(y)
df_Y[0].value_counts().plot(kind='bar')
plt.show()
```

```python
# 독립변수의 정규성 확인
import seaborn as sns
for i in range(df_X.shape[1]):
    sns.distplot(df_X[i])
    plt.title(feature_names[i])
    plt.show()
```

### PCA(주성분 분석 : Principal Component Analysis)

```python
# sklearn을 통해서 PCA 객체 생성 및 PC값을 구할 수 있습니다.
pca = PCA()
pca.fit(X)
PC_score = pca.transform(X)
PC_score[:5]
```

```python
# pca의 eigen_vector(고유벡터)
pca.components_
```

# PC 값의 설명력 정도

```python
# PC 값의 설명력 확인 : explained_variacne
pca.explained_variance_
```

```python
# explained_variance를 비율로 확인
ratio = pca.explained_variance_ratio_
ratio
# => PC1이 92%의 설명력을 가지고, PC4가 0.5%의 설명력을 가짐
```

```python
# pie plot을 통해 설명력의 정도 확인
df_v = pd.DataFrame(ratio, index=['PC1','PC2','PC3','PC4'], columns=['V_ration'])
df_v.plot.pie(y='V_ration')
df_v
```

```python
# feature에 따른 분포 확인
plt.figure()
colors = ['navy', 'turquoise', 'darkorange']
lw = 2

for color, i, target_name in zip(colors, [0,1,2] , iris.target_names):
    plt.scatter(X[y == i, 0], X[y == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset')
```
