```python
# 22798 : 마지막 페이지, 안정화를 위해 22797 페이지부터 크롤링 시작
nujuk_characterName = []  # 유저명 누적 리스트

start_tm = time.time()

for page_num in tqdm(range(22797,22790,-1), desc="진행 상태 "):
    
    # 던파 거래게시판 페이지 오픈
    URL = 'https://df.nexon.com/df/community/trade?order=reg_date&order_type=DESC&page=' + str(page_num)
    driver = webdriver.Chrome(executable_path='C:/Users/User/Desktop/Git_Hub/Study/유저분석/chromedriver_win32_98/chromedriver')
    driver.implicitly_wait(3)
    driver.get(url=URL)
    
    # 3~23 : 한 페이지에서 가져올 수 있는 유저명 수
    for i in range(3,23):
    #     xpath = '//*[@id="container"]/div[2]/div[2]/table/tbody/tr[i]/td[2]/span'
        xpath = '//*[@id="container"]/div[2]/div[2]/table/tbody/tr[' + str(i) + ']/td[2]/span'
        id_input = driver.find_element_by_xpath(xpath)
        characterName = id_input.text
        nujuk_characterName.append(characterName)
    time.sleep(1)
    
    driver.close()
    
print('소요시간 : ',time.time() - start_tm)
```
